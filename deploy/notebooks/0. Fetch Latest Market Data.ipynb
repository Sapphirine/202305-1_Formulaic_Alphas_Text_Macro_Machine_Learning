# Pull latest market data

## Imports

```python
import pandas as pd

import yfinance as yf

from datetime import datetime, date, timedelta

from s3 import read_from_s3, write_to_s3
```

## Constants

```python
CURR_DATE = (datetime.utcnow().date() - timedelta(days=1))if datetime.utcnow().hour < 22 else date.today()
```

```python
tickers = read_from_s3("security_map_final.csv").new_ticker
tickers
```

# Fetch newest data for all tickers

```python
failed_tickers = []
for ticker in tickers:
    ohlc_df = yf.download(ticker, start=CURR_DATE)
    if ohlc_df.empty:
        failed_tickers.append(ticker)
    else:
        pass
        # write_to_s3(ohlc_df, f"market_data_yf/{CURR_DATE.year}/{CURR_DATE.month}/{CURR_DATE.day}/{ticker}.csv")
```

```python
failed_ticker_df = pd.DataFrame({"ticker": failed_tickers})
write_to_s3(failed_ticker_df, f"market_data_yf/{CURR_DATE.year}/{CURR_DATE.month}/{CURR_DATE.day}/failed_tickers.csv")
```
